```python
import pandas as pd
df_magsubs = pd.read_json('Magazine_Subscriptions.json', lines=True)
df_metamagsubs = pd.read_json('meta_Magazine_Subscriptions.json', lines=True)

df_magsubs.to_csv()
df_magsubs
```

```python
df_magsubs = df_magsubs[["overall", "vote", "verified", "asin", "reviewerName", "reviewText", "summary"]]
df_magsubs.dropna(subset=["overall", "vote", "verified", "asin", "reviewerName", "reviewText", "summary"], inplace=True)
df_magsubs.reset_index()
df_magsubs
```

```python
df_metamagsubs.to_csv(index=False)
df_metamagsubs = df_metamagsubs[["category", "description", "title", "also_buy", "brand", "also_view", "details", "main_cat", "asin"]]
df_metamagsubs.dropna(subset=["category", "description", "title", "also_buy", "brand", "also_view", "details", "main_cat", "asin"], inplace=True)
df_metamagsubs.reset_index()
pd.options.display.max_colwidth = 200
df_metamagsubs
```

```python
df_categories = df_metamagsubs.category
#series_categories.dropna()
df_categories.str.split(pat=",")
df_categories
```

```python
df_categories.apply(pd.Series).stack()
```

```python
df_categories = df_categories.explode('Category')
df_categories = df_categories.to_frame()
df_categories
```

```python
df_categories = df_categories[df_categories.category != "Magazine Subscriptions"]
```

```python
df_CatCount = df_categories.value_counts().to_frame('amount')
df_CatCount.reset_index(inplace=True)
df_CatCount
```
