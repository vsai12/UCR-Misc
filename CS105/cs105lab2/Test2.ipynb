```python
import pandas as pd
import numpy as np
df = pd.read_csv("ca_dds_expenditures.csv")
table1 = pd.pivot_table(df, values=['Expenditures'], index=['Ethnicity'], aggfunc=np.mean)
table1.plot.bar()
```

```python
table2 = pd.pivot_table(df, values=['Expenditures'], index=['Ethnicity','Age Cohort'], aggfunc=np.mean)
table2 = table2.loc[["Hispanic","White not Hispanic"]]
table2.plot.bar()
```
